```python
## time module and function##

import time

initial = time.time()
k = 0
while (k<15):
    print("This is harry bhai")
    k+=1
print("While loop ran in", time.time() - initial, "second")    

initial2 = time.time()
for i in range(45):
    print("This is Rishab bhai")
print("For loop run in", time.time() - initial2, "second")
```

```python
# use time.sleep function(it will print late how muvh time i will be provide ##

import time

initial = time.time()
k = 0
while (k<15):
    print("This is harry bhai")
    time.sleep(2)
    k+=1
print("While loop ran in", time.time() - initial, "second")    

initial2 = time.time()
for i in range(45):
    print("This is Rishab bhai")
print("For loop run in", time.time() - initial2, "second")

## current time print function
import time
localtime = time.asctime(time.localtime(time.time()))
print(localtime)
```
