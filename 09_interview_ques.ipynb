```python
# create the N-gram
def generate_ngrams(sentence, n):
    words = sentence.split()
    ngrams = []
    
    for i in range(len(words) - n + 1):
        ngram = tuple(words[i:i+n])
        ngrams.append(ngram)
        
    return ngrams

sentence = "I am a data scientist"

# Examples
print("Unigrams:", generate_ngrams(sentence, 1))
print("Bigrams:", generate_ngrams(sentence, 2))
print("Trigrams:", generate_ngrams(sentence, 3))
```

```python
### Write the program reverse function, using buit-in function and with out buitin funtion.
def reverse_string(str1):
    new_str = ''
    for i in range(len(str1)-1, -1, -1):
        new_str = new_str + str1[i]
    return new_str

string1 = 'rishabh'
print(reverse_string(string1))


def reverse_string(str1):
    str1 = ''.join(reversed(str1))
    return str1

string1 = 'rishabh'
print(reverse_string(string1))
```
