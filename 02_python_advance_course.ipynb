### *args and **kwargs

*args - Used to pass multiple values (tuple)                    
      - (Inside function → it becomes a tuple)  
      
**kwargs - Used to pass key-value pairs                  
         - (Inside function → it becomes a dictionary)

```python
def add_numbers(*args):
    print(args)

add_numbers(1, 2, 3)
```

```python
def funargs(*args): 
    for item in args:
        print(item)
    
har = ["harry", "Rohan", "skillf", "Hammad"]
funargs(*har)
```

```python
def student_info(**kwargs):
    print(kwargs)

student_info(name="Rishabh", age=25)
```

```python
def funargs(normal,*args, **kwargs):
    print(normal)
    for item in args:
        print(item)
    for key, value in kwargs.items():
        print(key,value)

normal = "This is Rishabh"
har = ["Rishabh", "Rohan", "skillf", "Hammad"]
kw = {"rohan" : 12, "sanjay" : 13, "ruchi" : 14 }
funargs(normal,*har, **kw)
```

### Iterator

* **Iterator fetch one value at a time**                                        
If we have 100 values and we want to fetch one by one value use iterator

```python
name= "Rishabhvishwakarma"
vals = iter(name)
print(vals.__next__())
print(vals.__next__())
print(next(vals))
print(next(vals))
```

```python
class topten:
    def __init__(self):
        self.num = 1
    def __iter__(self):
        return self
    def __next__(self):
        val = self.num
        self.num += 1
        return val
values = topten()
print(values.__next__())
print(values.__next__())
```

#### 1. Duck Typing
👉 Python doesn’t care about type          
✔️ Only cares: does object have execute()?

```python
class VSCode:
    def execute(self):
        print("Running in VSCode")

class PyCharm:
    def execute(self):
        print("Running in PyCharm")

class Laptop:
    def code(self, ide):
        ide.execute()


lap = Laptop()
lap.code(VSCode())
lap.code(PyCharm())
```

## Multi Threading
Multithreading in Python means running multiple threads (smaller units of a process) at the same time within a single program.

One program → multiple tasks running concurrently

```python
from time import sleep
from threading import *

class tablet(Thread):
    def run(self):
        for i in range(5):
            print("hello")
            sleep(1)
            
class laptop(Thread):
    def run(self):
        for i in range(5):
            print("Hi")
            sleep(1)
            
t1 = tablet()
t2 = laptop()

t1.start()
sleep(0.2)
t2.start()

t1.join()
t2.join()
print("Bye")
```

### Getter and setter

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name 
        self.salary = salary

    def first_name(self):
        l = self.name.split(" ") 
        return l[0]
    
    def set_first_name(self, first):
        l = self.name.split(" ")
        new_name = f"{first} {l[1]}" 
        self.name = new_name

e = Employee("Jack Doe", 34555)
print(e.first_name())
e.set_first_name("John")
print(e.name)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name 
        self.salary = salary

    @property
    def first_name(self):
        l = self.name.split(" ") 
        return l[0]
    
    @first_name.setter
    def first_name(self, first):
        l = self.name.split(" ")
        new_name = f"{first} {l[1]}" 
        self.name = new_name

e = Employee("Jack Doe", 34555)
# print(e.first_name())
# e.set_first_name("John")
# print(e.name)

print(e.first_name)
e.first_name = "John"
print(e.name)
```

### Generator
A generator is the function that uses yield

👉 yield is used to generate values one at a time instead of returning everything at once.


```python
def nums():
    for i in range(20):
        yield i

print(map(str, nums()))
print(','.join(map(str, nums())))
```

```python
def divisible_by_5_and_7(n):
    for num in range(n + 1):
        if num % 5 == 0 and num % 7 == 0:
            yield num

try:
    n = int(input("Enter a value for n: "))
    result = divisible_by_5_and_7(n)
    print(','.join(map(str, result)))   # note here map use to convert interger to string
except ValueError:
    print("Invalid input. Please enter a valid integer for n.")  
```


## public, protect, private


#### Public
* Accessible everywhere
* Inside class, outside class, child class

```python
class Student:
    def __init__(self):
        self.name = "Rishabh"   # public

s = Student()
print(s.name)   # ✅ works
```

#### Protected
* Indicated by single underscore _
* Meant to be used inside class and child class
* Still accessible outside (but not recommended)

```python
class Student:
    def __init__(self):
        self._marks = 90   # protected


class Child(Student):
    def show(self):
        print(self._marks)   # ✅ allowed

s = Child()
s.show()

print(s._marks)  # ⚠️ works, but not recommended
```

#### Private

* Indicated by double underscore __
* Cannot be accessed directly outside the class

```python
class Student:
    def __init__(self):
        self.__marks = 95   # private

s = Student()

# print(s.__marks) ❌ error
```
