```python
# Write a program to find histogram of a given set of numbers. Take bin size from user. 
# Print the result in the form of a dictionary.

def histogram(dataset, bin_size):
    hist = {}

    for num in dataset:
        bin_start = (num // bin_size) * bin_size
        bin_end = bin_start + bin_size

        key = f"{bin_start}-{bin_end}"
        
        if key in hist:
            hist[key] =  hist[key] + 1
        else:
            hist[key] = 1

    return hist
        
# Example
data = [2, 5, 7, 10, 12, 14, 18, 21]
bin_size = int(input("Enter bin size: "))

result = histogram(data, bin_size)
print(result)
```

```python
# Write a program that accepts neighbors(set of 2D co-ordinates) and a point(single 2D co-ordinate) 
# and tells nearest neighbor(in terms of euclidean distance)

import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def nearest_neighbour(neighbors, point):
    neearest_dis = float('inf')
    nearest_point = None
    
    for x in neighbors:
        distance_cal = distance(x, point)
        if distance_cal < neearest_dis:
            neearest_dis = distance_cal
            nearest_point = x
            
    return neearest_dis, nearest_point

neighbors = [(1, 2), (4, 6), (5, 1), (2, 3)]
point = (3, 4)

n1, y = nearest_neighbour(neighbors, point)
print(n1)
print(y)
```

```python
# Write a program to swap the key value pair for max and min values
# if the dict is like this: {‘a’:1, ’b’:2,’c’:3}
# Output should be: {a:3,b:2,c:1}

def swap_dict(dict1):
    new_dict = {}
    min_values = min(dict1.values())
    max_values = max(dict1.values())

    for keys, values in dict1.items():
        if values == min_values:
            new_dict[keys] = max_values
        elif values == max_values:
            new_dict[keys] = min_values
        else:
            new_dict[keys] = values
    return new_dict
            
            
dict1 = {'a':1, 'b':2, 'c':3}
swap_dict(dict1)

    
```

```python
# Write a program that can sort a given unsorted list. Don't use any built-in function for sorting.

def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] =  arr[j + 1], arr[j]
    return arr

# Example
data = [5, 2, 9, 1, 5, 6, 3]
sorted_data = bubble_sort(data)

print(sorted_data)
```

```python
# 2. Given an array arr[] of size n, find the element that appears more than ⌊n/2⌋ times. If no such element exists, return -1.
# Input: arr[] = [1, 1, 2, 1, 3, 5, 1]
# Output: 1
# Explanation: Element 1 appears 4 times. Since ⌊7/2⌋ = 3, and 4 > 3, it is the majority element.

def majority_votes(arr):
    data = set(arr)
    new_dict = {}
    for x in arr:
        if x in data:
            new_dict[x] = new_dict.get(x, 0) + 1
            
    maximum = max(new_dict.values())
    element = len(arr)//2
    
    if maximum > element:
        return True
    else:
        return False
    
arr = [1,2,3,1,1,1,2]
print(majority_votes(arr))
arr = [2, 13]
print(majority_votes(arr))
```

```python
# 3. Find the number occuring odd number of times
# Given an array arr[] of positive integers. All numbers occur an even number of times except one 
# number which occurs an odd number of times.

def odd_number(arr):
    dict1 = {}
    for x in arr:
        dict1[x] = dict1.get(x, 0) + 1
        
    data = dict1.values()
    for x in data:
        if x%2 != 0:
            return x

arr = [1, 2, 3, 2, 3, 1, 3]
print(odd_number(arr))
```

```python
# 5 Write a Python Program for array rotation.
# Input:  arr[] = {1, 2, 3, 4, 5, 6, 7}, d = 2
# Output: 3, 4, 5, 6, 7, 1, 2
# Explanation: If the array is rotated by 1 position to the left, it becomes {2, 3, 4, 5, 6, 7, 1}.

def rotate_left(arr, k):
    k = k % len(arr)  # handle k > length
    return arr[k:] + arr[:k]

# Example
arr = [1, 2, 3, 4, 5]
print(rotate_left(arr, 2))  # Output: [3, 4, 5, 1, 2]
```

```python
# 6. Given an array arr[] of size n, the task is to find all the Leaders in the array. 
# An element is a Leader if it is greater than or equal to all the elements to its right side.

def leaders(arr):
    result = []
    n = len(arr)
    
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] < arr[j]:
                break
        else:
            # If no larger element was found
            result.append(arr[i])
    
    return result

arr = [16, 17, 4, 3, 5, 2]
result = leaders(arr)
print(" ".join(map(str, result)))
```

```python
# 9. Product of Array Except Self
# Input: arr[] = [10, 3, 5, 6, 2]
# Output: [180, 600, 360, 300, 900]
# Explanation: 
# For i=0, res[i] = 3 * 5 * 6 * 2 is 180.
# For i = 1, res[i] = 10 * 5 * 6 * 2 is 600.
# For i = 2, res[i] = 10 * 3 * 6 * 2 is 360.
# For i = 3, res[i] = 10 * 3 * 5 * 2 is 300.
# For i = 4, res[i] = 10 * 3 * 5 * 6 is 900.

def multiply(arr1):
    new_list1 = []
    for x in range(len(arr1)):
        value1 = 1
        for y in range(len(arr1)):
            if x ==y:
                continue
            else:
                value1 = value1 * arr1[y]
        new_list1.append(value1)
        
    return new_list1
            
arr = [10, 3, 5, 6, 2]
print(multiply(arr))
```

```python
# 13. Subarray with Given Sum
# Input: arr[] = [15, 2, 4, 8, 9, 5, 10, 23], target = 23
# Output: [2, 5]
# Explanation: Sum of subarray arr[2...5] is 2 + 4 + 8 + 9 = 23.

def get_data(arr1, target):
    for i in range(len(arr1)):
        sum_value = 0
        store_value = []
        for j in range(i, len(arr1)):
            sum_value = sum_value + arr1[j]
            store_value.append(arr1[j])
            if sum_value == target:
                return store_value
    return store_value
                
arr1 = [15, 2, 4, 8, 9, 5, 10, 23]
print(get_data(arr1, 23))
```

```python
# 14. Given an unsorted array arr[] with both positive and negative elements, find the smallest positive number missing from the array.
# Input: arr[] = [2, -3, 4, 1, 1, 7]
# Output: 3
# Explanation: 3 is the smallest positive number missing from the array.

def get_data(arr1):
    data = sorted(arr1)
    max_value = max(data)
    
    for x in range(max_value):
        if x not in data and x > 0:
            return x
            break
        
    return x
arr1 = [2, -3, 4, 1, 1, 7]
print(get_data(arr1))
```

```python
# 16. Given an array arr[] containing only 0s and 1s, find the longest subarray which contains equal no of 0s and 1s.
# Input: arr[] = [1, 0, 1, 1, 1, 0, 0]
# Output: 6
# Explanation: arr[1 ... 6] is the longest subarray with three 0s and three 1s.

def longest_subarray(arr):
    max_length = 0
    for i in range(len(arr)):
        count_0 = 0
        count_1 = 0
        for j in range(i, len(arr)):
            if arr[j] == 0:
                count_0 += 1
            else:
                count_1 += 1
            if count_0 == count_1:
                max_length = max(max_length, j - i + 1)
    return max_length

arr = [1, 0, 1, 1, 1, 0, 0]
print(longest_subarray(arr))
```
