# String

```python
def char_frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
char_frequency('google.com')
```

```python
def add_string(str):
    length = len(str)
    if length>2:
        if str[-3:] == "ing":
            str += "ly"
        else:
            str += "ing"
    return str

print(add_string("string"))
print(add_string("school")) 
```

```python
# 8.Write a Python function that takes a list of words and return the longest word and the length of the longest one

def find_longest_word(words_list):
    word_len = []
    for x in words_list:
        word_len.append((len(x),x))
    word_len.sort()
    return word_len

find_longest_word(["PHP", "Exercises", "Backend"])
```

```python
def rev_str(str):
    if len(str)%4==0:
        return ''.join(reversed(str))
    return str
rev_str("abcd")
```

# List [ ]

```python
def sum_value(lst):
    sum = 0
    for x in lst:
        sum += x
    return sum
sum_value([2,3,4,5,6])
    
```

```python
def mul_value(lst):
    sum = 1
    for x in lst:
        sum = sum * x
    return sum
mul_value([2,3,4,5,6])
```

```python
def smallest_value(ls):
    value = ls[0]
    for x in ls:
        if x < value:
            value = x
    return value
smallest_value([2,30,4,5,6,24])
```

```python
def largest_value(ls):
    value = ls[0]
    for x in ls:
        if x > value:
            value = x
    return value
largest_value([2,30,4,5,6,24])
```

```python
def count_str(ls):
    count = 0
    for x in ls:
        if len(x) > 1 and x[0]==x[-1]:
            count +=1
    return count
count_str(['abc','xyz','aba','a22a'])       
```

```python
# 37. Write a Python program to find common items from two lists

def common_list(ls1, ls2):
    list3 = []
    for x in ls1:
        for y in ls2:
            if x==y:
                list3.append(x)
    return list3
common_list([2,4,6,8,9],[2,5,6,8,1])
```

```python
# 45. Write a Python program to convert a pair of values into a sorted unique array.

L = [(1, 2), (3, 4), (1, 2), (5, 6), (7, 8), (1, 2), (3, 4), (3, 4), (7, 8), (9, 10)]
l1 = []
for x in L:
    for y in x:
        if y not in l1:
            l1.append(y)
l1.sort()
print(l1)
```
