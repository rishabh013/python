# String

```python
def char_frequency(str1):
    dict = {}
    for n in str1:
        keys = dict.keys()
        if n in keys:
            dict[n] += 1
        else:
            dict[n] = 1
    return dict
char_frequency('google.com')
```

```python
def add_string(str):
    length = len(str)
    if length>2:
        if str[-3:] == "ing":
            str += "ly"
        else:
            str += "ing"
    return str

print(add_string("string"))
print(add_string("school")) 
```

```python
# 8.Write a Python function that takes a list of words and return the longest word and the length of the longest one


def find_longest_word(list1):
    new_num = 0
    word = ''
    
    for x in list1:
        if len(x) > new_num:
            new_num = len(x)
            word = x
    
    return word, new_num
print(find_longest_word(["PHP", "Exercises", "Backend"]))

def find_longest_word(list1):
    dict1 = {}
    for x in list1:
        dict1[x] = len(x)
    
    new_num = 0
    word = ''
    for key , value in dict1.items():
        if new_num < value:
            new_num = value
            word = key
    return word, new_num

print(find_longest_word(["PHP", "Exercises", "Backend"]))
```

```python
def rev_str(str):
    if len(str)%4==0:
        return ''.join(reversed(str))
    return str
rev_str("abcd")
```

# List [ ]

```python
def smallest_value(ls):
    value = ls[0]
    for x in ls:
        if x < value:
            value = x
    return value
smallest_value([2,30,4,5,6,24])
```

```python
def largest_value(ls):
    value = ls[0]
    for x in ls:
        if x > value:
            value = x
    return value
largest_value([2,30,4,5,6,24])
```

```python
def count_str(ls):
    count = 0
    for x in ls:
        if len(x) > 1 and x[0]==x[-1]:
            count +=1
    return count
count_str(['abc','xyz','aba','a22a'])       
```

```python
# 37. Write a Python program to find common items from two lists

def common_list(ls1, ls2):
    list3 = []
    for x in ls1:
        for y in ls2:
            if x==y:
                list3.append(x)
    return list3
common_list([2,4,6,8,9],[2,5,6,8,1])
```

```python
# 42. Write a Python program to find missing and additional values in two lists.

list1 = ['a','b','c','d','e','f']
list2 = ['d','e','f','g','h']
print('Missing values in second list: ', ','.join(set(list1).difference(list2)))
print('Additional values in second list: ', ','.join(set(list2).difference(list1)))
```
