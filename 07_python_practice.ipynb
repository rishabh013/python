```python
# create the N-gram
def generate_ngrams(sentence, n):
    words = sentence.split()
    ngrams = []
    
    for i in range(len(words) - n + 1):
        ngram = tuple(words[i:i+n])
        ngrams.append(ngram)
        
    return ngrams

sentence = "I am a data scientist"

# Examples
print("Unigrams:", generate_ngrams(sentence, 1))
print("Bigrams:", generate_ngrams(sentence, 2))
print("Trigrams:", generate_ngrams(sentence, 3))
```

```python
import threading
import time

def task(name):
    print(f"Starting {name}")
    time.sleep(2)  # Simulate I/O task
    print(f"Finished {name}")

# Create threads
t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))

# Start threads
t1.start()
t2.start()

# Wait for completion
t1.join()
t2.join()

print("All threads completed")
```

```python
# 14. Write a Python Program to check a prime number.
def is_prime(num):
    if num <= 1:
        return False
        
    elif num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                return False
    return True
            
is_prime(6)
```

```python
# 15. Write a Python Program to Print all Prime Numbers in an Interval of 1-10
def is_prime(num):
    if num <= 1:
        return False
        
    elif num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                return False
    return True


def primes_upto_n(N):
    for num in range(2, N + 1):
        if is_prime(num):
            print(num, end=" ")

primes_upto_n(20)
```

```python
# 36. Write a Python Program to check if the given array is Monotonic.
# Answer : A monotonic array is one that is entirely non-increasing or non-decreasing

def is_monotonic(arr):
    increasing = True
    decreasing = True

    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            increasing = False
        if arr[i] < arr[i + 1]:
            decreasing = False

    return increasing or decreasing


# Example
print(is_monotonic([1,2,2,3]))     # True
print(is_monotonic([3,2,1]))       # True
print(is_monotonic([1, 3, 2, 4]))  # Faalse
```

```python
# 37. Write a Python Program to Add Two Matrices.

def add_matrices(A, B):
    result = []
    for i in range(len(A)):             # rows
        row = []
        for j in range(len(A[0])):      # columns
            row.append(A[i][j] + B[i][j])
        result.append(row)

    return result


A = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

B = [[9, 8, 7],
     [6, 5, 4],
     [3, 2, 1]]

result = add_matrices(A, B)

# Print result
for row in result:
    print(row)
```

```python
# 38 Write a Python Program to Multiply Two Matrices

def multiply_matrices(mat1, mat2):
    # Determine dimensions
    rows1 = len(mat1)
    cols1 = len(mat1[0])
    rows2 = len(mat2)
    cols2 = len(mat2[0])

    # Check if multiplication is possible
    if cols1 != rows2:
        return "Matrix multiplication is not possible."

    # Initialize result matrix
    result = [[0 for _ in range(cols2)] for _ in range(rows1)]

    # Matrix multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += mat1[i][k] * mat2[k][j]

    return result


# Example matrices
matrix1 = [
    [1, 2, 3],
    [4, 5, 6]
]

matrix2 = [
    [7, 8],
    [9, 10],
    [11, 12]
]

# Multiply matrices
result_matrix = multiply_matrices(matrix1, matrix2)

# Print result
if isinstance(result_matrix, str):
    print(result_matrix)
else:
    print("Result of matrix multiplication:")
    for row in result_matrix:
        print(row)
```

```python
# 79 Write a program that accepts a sentence and calculates the number of letters and digits. 
# Suppose the following input is supplied to the program: ex - "hello world! 123"

def word_count(words):
    letter_count = 0
    digit_count = 0
    for char in words:
        if char.isalpha():
            letter_count += 1
        elif char.isdigit():
            digit_count +=1

    return letter_count, digit_count
    
words = "hello world! 123"
word_count(words)
```

```python
# 86. Please write a binary search function which searches an item in a sorted list.
#     The function should return the index of the element to be searched in the list.

def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid                # Element found, return index
        elif arr[mid] < target:
            left = mid + 1            # Search in right half
        else:
            right = mid - 1           # Search in left half

    return -1  # Element not found

arr = [2, 5, 8, 12, 16, 23, 38]
target = 16

result = binary_search(arr, target)

if result != -1:
    print(f"Element found at index: {result}")
else:
    print(f"Element not found")
```

```python
# 87. Please write a program using a generator to print the numbers which can be divisible
#     by 5 and 7 between 0 and n in comma-separated form, while n is input by console.

def divisible_by_5_and_7(n):
    for num in range(n + 1):
        if num % 5 == 0 and num % 7 == 0:
            yield num

try:
    n = int(input("Enter a value for n: "))
    result = divisible_by_5_and_7(n)
    print(','.join(map(str, result)))   # note here map use to convert interger to string
except ValueError:
    print("Invalid input. Please enter a valid integer for n.")    
```

```python
# 98. Create a function that returns True if a given inequality expression is correct and False otherwise.

def correct_signs(str1):
    word = str1.split()

    for i in range(1, len(word), 2):
        left = int(word[i - 1])
        operator = word[i]
        right = int(word[i + 1])

        if operator == '<' and not (left < right):
            return False
        if operator == '>' and not (left > right):
            return False

    return True


print(correct_signs("1 < 2 < 6 < 9 > 3"))  # True
print(correct_signs("13 > 44 > 33 < 1"))    # False
```

```python
# 125 An isogram is a word that has no duplicate letters. Create a function that takes a
# string and returns either True or False depending on whether or not it's an "isogram".

def is_isogram(string1):
    string2 = string1.lower()
    new_dict = set()

    for x in string2:
        if x in new_dict:
            return False
        new_dict.add(x)
    return True
    
print(is_isogram("PasSword"))
is_isogram("Algorism")
```

```python
# 133 Create a function that takes a list of numbers and return the number that's unique.
def unique(string1):
    new_dict = {}
    for x in string1:
        if x in new_dict:
            new_dict[x] = new_dict[x] +1
        else:
            new_dict[x] = 1
            
    for key, values in new_dict.items():
        if values == 1:
            return key
    
print(unique([3, 3, 3, 7, 3, 3]))
unique([0, 1, 1, 1, 1, 1, 1, 1])
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def getArea(self):
        # Calculate and return the area of the circle
        return round(math.pi * self.radius**2)

    def getPerimeter(self):
        # Calculate and return the perimeter (circumference) of the circle
        return round(2 * math.pi * self.radius)


# Test cases
circy = Circle(11)
print(circy.getArea())
print(circy.getPerimeter())
```

```python
# 135 Create a function that takes a list of strings and return a list, sorted from shortest to longest.
# sort_by_length(["Google", "Apple", "Microsoft"]) ➞ ["Apple", "Google", "Microsoft"]

def sort_by_length(list1):
    
    dict = {}
    for x in list1:
        dict[x] = len(x)
        
    new_list = []
    for y in sorted(dict.values()):
        for key, value in dict.items():
            if y == value:
                new_list.append(key)
        
    return new_list
    
sort_by_length(["Google", "Apple", "Microsoft"])
```
