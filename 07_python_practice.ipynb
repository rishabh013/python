```python
# create the N-gram
def generate_ngrams(sentence, n):
    words = sentence.split()
    ngrams = []
    
    for i in range(len(words) - n + 1):
        ngram = tuple(words[i:i+n])
        ngrams.append(ngram)
        
    return ngrams

sentence = "I am a data scientist"

# Examples
print("Unigrams:", generate_ngrams(sentence, 1))
print("Bigrams:", generate_ngrams(sentence, 2))
print("Trigrams:", generate_ngrams(sentence, 3))
```

```python
import threading
import time

def task(name):
    print(f"Starting {name}")
    time.sleep(2)  # Simulate I/O task
    print(f"Finished {name}")

# Create threads
t1 = threading.Thread(target=task, args=("Thread-1",))
t2 = threading.Thread(target=task, args=("Thread-2",))

# Start threads
t1.start()
t2.start()

# Wait for completion
t1.join()
t2.join()

print("All threads completed")
```

```python
# Write a Python program of recursion list sum.
def sum_list(list):
    total = 0
    for element in list:
        if type(element) == type([]):
             total = total + sum_list(element)
                
        else:
            total = total + element
    return total  
print(sum_list([1, 2, [1,2], [3,4]]))   
```

```python
# 79 Write a program that accepts a sentence and calculates the number of letters and digits. 
# Suppose the following input is supplied to the program: ex - "hello world! 123"

def word_count(words):
    letter_count = 0
    digit_count = 0
    for char in words:
        if char.isalpha():
            letter_count += 1
        elif char.isdigit():
            digit_count +=1

    return letter_count, digit_count
    
words = "hello world! 123"
word_count(words)
```

```python
# Please write a program using a generator to print the numbers which can be divisible
# by 5 and 7 between 0 and n in comma-separated form, while n is input by console.
def divisible_by_5_and_7(n):
    for num in range(n + 1):
        if num % 5 == 0 and num % 7 == 0:
            yield num

try:
    n = int(input("Enter a value for n: "))
    result = divisible_by_5_and_7(n)
    print(','.join(map(str, result)))   # note here map use to convert interger to string
except ValueError:
    print("Invalid input. Please enter a valid integer for n.")            
```

```python
# 100. Write a function that calculates the factorial of a number recursively.

def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
    
print(fact(5))
```

```python
# 101. Create a function that computes the Hamming distance between two strings.
# Examples
# hamming_distance("abcde", "bcdef") ➞ 5
# hamming_distance("abcde", "abcde") ➞ 0
# hamming_distance("strong", "strung") ➞ 1

def hamming(str1, str2):
    count1 = 0
    for x, y in zip(str1, str2):
        if x ==y:
            pass
        else:
            count1 += 1
    return count1
    
print(hamming("abcde", "bcdef"))
print(hamming("strong", "strung"))
```

```python
# 110 Create a function that takes a single string as argument and returns an ordered list 
# containing the indices of all capital letters in the string.

def index_of_caps(str1):
    new_list = []
    for index, x in enumerate(str1):
        if x.isupper():
            new_list.append(index)
    return new_list

print(index_of_caps("eDaBiT"))

# use list comprehensions
def index_of_caps(str1):
    return [index for index, x in enumerate(str1) if x.isupper()]

print(index_of_caps("aDjaDW"))
```

```python
# 125 An isogram is a word that has no duplicate letters. Create a function that takes a
# string and returns either True or False depending on whether or not it's an "isogram".

def is_isogram(string1):
    string2 = string1.lower()
    new_dict = set()

    for x in string2:
        if x in new_dict:
            return False
        new_dict.add(x)
    return True
    
print(is_isogram("PasSword"))
is_isogram("Algorism")
```

```python
# 133 Create a function that takes a list of numbers and return the number that's unique.
def unique(string1):
    new_dict = {}
    for x in string1:
        if x in new_dict:
            new_dict[x] = new_dict[x] +1
        else:
            new_dict[x] = 1
            
    for key, values in new_dict.items():
        if values == 1:
            return key
    
print(unique([3, 3, 3, 7, 3, 3]))
unique([0, 1, 1, 1, 1, 1, 1, 1])
```

```python
def unique(string1):
    new_dict = {}
    
    # Count frequency
    for x in string1:
        new_dict[x] = new_dict.get(x, 0) + 1
    
    # Find element with frequency 1
    for key, value in new_dict.items():
        if value == 1:
            return key

print(unique([3, 3, 3, 7, 3, 3]))
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def getArea(self):
        # Calculate and return the area of the circle
        return round(math.pi * self.radius**2)

    def getPerimeter(self):
        # Calculate and return the perimeter (circumference) of the circle
        return round(2 * math.pi * self.radius)


# Test cases
circy = Circle(11)
print(circy.getArea())
print(circy.getPerimeter())
```

```python
# 135 Create a function that takes a list of strings and return a list, sorted from shortest to longest.
# sort_by_length(["Google", "Apple", "Microsoft"]) ➞ ["Apple", "Google", "Microsoft"]

def sort_by_length(list1):
    
    dict = {}
    for x in list1:
        dict[x] = len(x)
        
    new_list = []
    for y in sorted(dict.values()):
        for key, value in dict.items():
            if y == value:
                new_list.append(key)
        
    return new_list
    
sort_by_length(["Google", "Apple", "Microsoft"])
```
